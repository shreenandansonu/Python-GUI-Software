# Detailed Notes on Linear Time-Invariant (LTI) Systems and RC Circuits

This document provides a comprehensive summary of the lecture on Linear Time-Invariant (LTI) systems, focusing on convolution, eigenfunctions, Laplace transforms, transfer functions, frequency response, and the behavior of first-order systems, particularly RC circuits. The notes include explanations, mathematical derivations, and relevant diagrams to aid understanding.

## 1. Convolution in LTI Systems

### Concept
Convolution is a fundamental operation in LTI systems used to compute the output of a system given its **impulse response** and an arbitrary input.

- **Impulse Response**: The output of an LTI system when the input is an impulse function (Dirac delta function, δ(t)). Denoted as **h(t)**, it fully characterizes the system's input-output relationship.
- **Convolution Operation**: For an input **x(t)** and impulse response **h(t)**, the output **y(t)** is given by the convolution integral:
  \[
  y(t) = x(t) * h(t) = \int_{-\infty}^{\infty} x(\tau) h(t - \tau) d\tau
  \]
  This integral slides the flipped impulse response over the input signal, computing the output at each time **t**.

### Significance
- If you know **h(t)**, you can compute the output for any input **x(t)** by convolving it with **h(t)**.
- Convolution captures how the system responds to any input by summing the contributions of the impulse response scaled by the input.

## 2. Eigenfunctions of LTI Systems

### Definition
An **eigenfunction** of a system is an input function that, when applied, produces an output that is a scaled version of itself. For LTI systems, the eigenfunctions are **complex exponential signals** of the form:
\[
x(t) = e^{st}, \quad s = \sigma + j\omega
\]
where **s** is a complex number, **σ** is the real part, and **ω** is the imaginary part.

### Property
When an exponential signal **e^{st}** is input to an LTI system, the output is:
\[
y(t) = H(s) e^{st}
\]
- **H(s)** is a complex number called the **eigenvalue** associated with the eigenfunction **e^{st}**.
- **H(s)** is the **transfer function** evaluated at **s**, which we will discuss later.

### Analogy to Linear Algebra
- In linear algebra, an eigenvector **x** of a matrix **A** satisfies:
  \[
  A x = \lambda x
  \]
  where **λ** is the eigenvalue.
- In LTI systems, **e^{st}** is analogous to the eigenvector, and **H(s)** is analogous to the eigenvalue.

## 3. Laplace Transform

### Definition
The Laplace transform converts a time-domain signal into the s-domain, facilitating the analysis of LTI systems, especially for solving differential equations.

- **Unilateral Laplace Transform**:
  \[
  X(s) = \int_{0^-}^{\infty} x(t) e^{-st} dt
  \]
  - Used for causal systems (signals starting at **t = 0**).
  - Assumes the signal is zero for **t < 0**.

- **Bilateral Laplace Transform**:
  \[
  X(s) = \int_{-\infty}^{\infty} x(t) e^{-st} dt
  \]
  - Considers the entire time axis, suitable for non-causal systems.

### Application
- The Laplace transform simplifies the solution of **linear constant coefficient differential equations** by converting them into algebraic equations in the s-domain.
- For a differential equation, the Laplace transform of the derivative **dx(t)/dt** is:
  \[
  \mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = s X(s) - x(0^-)
  \]
  Assuming initial conditions (e.g., **x(0^-) = 0**) simplifies the analysis.

## 4. Transfer Function

### Definition
The **transfer function** **H(s)** of an LTI system is defined as the ratio of the Laplace transform of the output **Y(s)** to the Laplace transform of the input **X(s)**, assuming zero initial conditions:
\[
H(s) = \frac{Y(s)}{X(s)}, \quad X(s) \neq 0
\]
Alternatively, **H(s)** is the Laplace transform of the impulse response **h(t)**:
\[
H(s) = \mathcal{L}\{h(t)\} = \int_{0^-}^{\infty} h(t) e^{-st} dt
\]

### Significance
- **H(s)** encapsulates the system's behavior in the s-domain.
- The output in the s-domain is:
  \[
  Y(s) = H(s) X(s)
  \]
- To find the time-domain output **y(t)**, take the inverse Laplace transform:
  \[
  y(t) = \mathcal{L}^{-1}\{H(s) X(s)\}
  \]

## 5. Frequency Response

### Definition
The **frequency response** is obtained by evaluating the transfer function **H(s)** at **s = jω**, where **ω** is the angular frequency:
\[
H(j\omega) = H(s)|_{s=j\omega}
\]
- **H(jω)** is a complex number with a magnitude and phase:
  \[
  H(j\omega) = |H(j\omega)| e^{j \theta(\omega)}
  \]
  - **|H(jω)|**: Magnitude response (gain).
  - **θ(ω)**: Phase response (phase shift).

### Interpretation
- If a sinusoidal input **x(t) = sin(ωt)** is applied, the output is:
  \[
  y(t) = |H(j\omega)| \sin(\omega t + \theta(\omega))
  \]
- The magnitude **|H(jω)|** determines the amplification or attenuation of the input amplitude.
- The phase **θ(ω)** indicates the phase shift of the output relative to the input.
- For LTI systems, the frequency **ω** remains unchanged; only amplitude and phase are modified.

### Representation
- **Magnitude Response** in decibels (dB):
  \[
  |H(j\omega)|_{\text{dB}} = 20 \log_{10} |H(j\omega)|
  \]
- **Phase Response** in radians or degrees (converted for readability).

## 6. First-Order Systems

### Definition
A first-order system is described by a first-order differential equation:
\[
a_1 \frac{dy(t)}{dt} + a_0 y(t) = b_0 x(t)
\]
Dividing by **a_0**:
\[
\tau \frac{dy(t)}{dt} + y(t) = k x(t)
\]
- **τ = a_1 / a_0**: Time constant (determines the rate of response).
- **k = b_0 / a_0**: Static sensitivity (steady-state gain).

### Impulse Response
The impulse response **h(t)** of a first-order system is:
\[
h(t) = \frac{k}{\tau} e^{-t/\tau} u(t)
\]
- **u(t)**: Unit step function (ensures **h(t) = 0** for **t < 0**).
- **k/τ**: Initial amplitude at **t = 0^+**.
- The response is an exponentially decaying function for **t ≥ 0**.

#### Effect of Parameters
- **k**: Scales the amplitude of the response. Larger **k** increases the initial height.
- **τ**: Controls the decay rate. Larger **τ** slows the decay, making the response more horizontal; smaller **τ** speeds up the decay.

### Example: RC Circuit

#### Circuit Diagram
Consider an RC circuit with a resistor **R** and capacitor **C** in series, with input voltage **v_in(t)** and output voltage **v_out(t)** across the capacitor.

![RC Circuit Diagram](https://i.imgur.com/rc_circuit_diagram.png)
*Figure 2: RC circuit with input v_in(t) and output v_out(t) across the capacitor.*

#### Differential Equation
Using Kirchhoff’s Voltage Law (KVL):
\[
v_{in}(t) = v_R(t) + v_{out}(t)
\]
- Voltage across resistor: **v_R(t) = i(t) R**.
- Voltage across capacitor: **v_{out}(t)**.
- Current through capacitor: **i(t) = C \frac{d v_{out}(t)}{dt}**.

Substitute **i(t)**:
\[
v_{in}(t) = R C \frac{d v_{out}(t)}{dt} + v_{out}(t)
\]
This is a first-order differential equation:
\[
RC \frac{d v_{out}(t)}{dt} + v_{out}(t) = v_{in}(t)
\]
- Time constant: **τ = RC**.
- Static sensitivity: **k = 1**.

#### Impulse Response
For an impulse input **v_{in}(t) = δ(t)**, the impulse response is:
\[
h(t) = \frac{1}{RC} e^{-t/RC} u(t)
\]
- Initial amplitude: **1/RC** at **t = 0^+**.
- Decays exponentially with time constant **RC**.

#### Laplace Domain Analysis
Taking the Laplace transform of the differential equation (assuming zero initial conditions):
\[
RC s V_{out}(s) + V_{out}(s) = V_{in}(s)
\]
\[
V_{out}(s) (RC s + 1) = V_{in}(s)
\]
\[
H(s) = \frac{V_{out}(s)}{V_{in}(s)} = \frac{1}{RC s + 1}
\]
Alternatively, using impedance:
- Resistor impedance: **Z_R = R**.
- Capacitor impedance: **Z_C = 1/(sC)**.
- The circuit acts as a voltage divider:
  \[
  V_{out}(s) = V_{in}(s) \cdot \frac{Z_C}{Z_R + Z_C} = V_{in}(s) \cdot \frac{1/sC}{R + 1/sC} = \frac{1}{RC s + 1} V_{in}(s)
  \]
- Transfer function: **H(s) = \frac{1}{RC s + 1}**.
- Impulse response: **h(t) = \mathcal{L}^{-1}\{H(s)\} = \frac{1}{RC} e^{-t/RC} u(t)**.

#### Frequency Response
Substitute **s = jω**:
\[
H(j\omega) = \frac{1}{RC j\omega + 1} = \frac{1}{1 + j\omega RC}
\]
- Magnitude:
  \[
  |H(j\omega)| = \frac{1}{\sqrt{1 + (\omega RC)^2}}
  \]
  In dB:
  \[
  |H(j\omega)|_{\text{dB}} = 20 \log_{10} \left( \frac{1}{\sqrt{1 + (\omega RC)^2}} \right)
  \]
- Phase:
  \[
  \theta(\omega) = -\tan^{-1}(\omega RC)
  \]
- **Low-Pass Filter**: The system passes low frequencies (small **ω**) with minimal attenuation and attenuates high frequencies (large **ω**).

![Frequency Response of RC Circuit](https://i.imgur.com/frequency_response_rc.png)
*Figure 3: Magnitude and phase response of an RC circuit (logarithmic frequency axis).*

#### Step Response
For a step input **v_{in}(t) = u(t)**:
\[
V_{in}(s) = \frac{1}{s}
\]
\[
V_{out}(s) = H(s) V_{in}(s) = \frac{1}{RC s + 1} \cdot \frac{1}{s}
\]
Using partial fraction expansion and inverse Laplace transform:
\[
v_{out}(t) = (1 - e^{-t/RC}) u(t)
\]
- Small **τ**: Quick rise to steady-state value (1).
- Large **τ**: Slower rise.

## 7. System Order and Input-Output Selection

### Key Insight
The order of a system (static or dynamic) depends on the choice of input and output variables.

- **Static System (Zero-Order)**: Output depends only on the current input, e.g., **y(t) = k x(t)**.
  - Example: Force input, displacement output in a spring system (**F = kx**).
- **First-Order System**: Involves first derivative, e.g., RC circuit or damper system with force input and position output.
- **Example Variations**:
  - **Mass on frictionless surface**:
    - Input: Force, Output: Acceleration → Static system (**F = ma**).
    - Input: Force, Output: Position → Second-order system (involves second derivative).

## 8. Second-Order Systems

### Definition
A second-order system is described by a second-order differential equation:
\[
a_2 \frac{d^2 y(t)}{dt^2} + a_1 \frac{dy(t)}{dt} + a_0 y(t) = b_0 x(t)
\]
Dividing by **a_0**:
\[
\frac{a_2}{a_0} \frac{d^2 y(t)}{dt^2} + \frac{a_1}{a_0} \frac{dy(t)}{dt} + y(t) = \frac{b_0}{a_0} x(t)
\]
Define:
- **Static sensitivity**: **k = b_0 / a_0**.
- **Natural frequency**: **ω_n = \sqrt{a_0 / a_2}**.
- **Damping ratio**: **ζ = \frac{a_1}{2 \sqrt{a_0 a_2}}**.

Standard form:
\[
\frac{1}{\omega_n^2} \frac{d^2 y(t)}{dt^2} + \frac{2ζ}{\omega_n} \frac{dy(t)}{dt} + y(t) = k x(t)
\]

### Transfer Function
Taking the Laplace transform (zero initial conditions):
\[
\left( \frac{s^2}{\omega_n^2} + \frac{2ζ s}{\omega_n} + 1 \right) Y(s) = k X(s)
\]
\[
H(s) = \frac{Y(s)}{X(s)} = \frac{k}{\frac{s^2}{\omega_n^2} + \frac{2ζ s}{\omega_n} + 1} = \frac{k \omega_n^2}{s^2 + 2ζ \omega_n s + \omega_n^2}
\]

### Impulse Response
The impulse response depends on **ζ**:
- **Undamped (ζ = 0)**: Oscillatory (sinusoidal) response.
- **Underdamped (0 < ζ < 1)**: Oscillatory with decaying amplitude.
- **Critically Damped (ζ = 1)**: Fastest non-oscillatory return to zero.
- **Overdamped (ζ > 1)**: Slow non-oscillatory return to zero.

### Example: RLC Circuit
An RLC circuit (resistor, inductor, capacitor) is a second-order system with two memory elements (inductor and capacitor).

![RLC Circuit Diagram](https://i.imgur.com/rlc_circuit_diagram.png)
*Figure 6: RLC circuit.*

### Frequency Response
\[
H(j\omega) = \frac{k \omega_n^2}{(j\omega)^2 + 2ζ \omega_n (j\omega) + \omega_n^2}
\]
- **Undamped**: Infinite peak at **ω = ω_n** (resonance).
- **Underdamped**: Peak near **ω_n**, height increases as **ζ** decreases.
- **Critically/Overdamped**: No peak, smoother roll-off.

![Frequency Response of Second-Order System](https://i.imgur.com/second_order_frequency.png)
*Figure 7: Magnitude response for different damping ratios.*

### Resonance
- Second-order systems exhibit **resonance** due to two memory elements (e.g., inductor and capacitor, or mass and spring) exchanging energy.
- Low damping (**ζ** small) leads to high gain at **ω_n**.

## 9. Key Takeaways
- **Convolution** and **impulse response** allow computation of LTI system outputs.
- **Eigenfunctions** (exponentials) simplify system analysis, with **H(s)** as eigenvalues.
- **Laplace transform** converts differential equations to algebraic form.
- **Transfer function** **H(s)** relates input and output in the s-domain.
- **Frequency response** **H(jω)** describes amplitude and phase changes for sinusoidal inputs.
- **First-order systems** (e.g., RC circuits) have exponential responses, acting as low-pass filters.
- **Second-order systems** (e.g., RLC circuits) exhibit complex behaviors (oscillatory, damped) due to two memory elements.
- **System order** depends on input-output selection.

## 10. Visual Summary

### RC Circuit Analysis Flow
1. **Time Domain**: Derive differential equation using KVL.
2. **Laplace Domain**: Compute **H(s)** using impedance or Laplace transform.
3. **Impulse Response**: Inverse Laplace transform of **H(s)**.
4. **Frequency Response**: Evaluate **H(jω)** for magnitude and phase.
5. **Step Response**: Compute for step input to analyze transient behavior.

### Plots
- **Impulse Response**: Shows transient behavior.
- **Frequency Response**: Shows steady-state behavior for sinusoids.
- **Step Response**: Shows transition to steady state.
